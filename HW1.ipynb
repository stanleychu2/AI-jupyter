```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = []
y = []
turn = -12
for i in range(0, 360):
    x.append(np.cos(i))
    y.append(np.random.randint(20) * np.sin(i) + np.random.randint(10) * np.tan(i) * turn)
    turn *= -1
plt.plot(x, y, color="brown")
```
